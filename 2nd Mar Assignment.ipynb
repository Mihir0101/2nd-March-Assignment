Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

--> A scatter plot is a type of plot that displays the relationship between two variables. It is a graphical representation     of the correlation between two variables

```python
import matplotlib.pyplot as plt

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.xlabel('I am X')
plt.ylabel('I am Y')
plt.title('I am Title')

plt.show()
```

Q3: Why is the subplot() function used? 
Draw four line plots using the subplot() function.

```python
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])


fig, axs = plt.subplots(2, 2)

axs[0,0].plot(x1,y1)
axs[0,1].plot(x2,y2)
axs[1,0].plot(x3,y3)
axs[1,1].plot(x4,y4)

axs[0, 0].set_title('1st')
axs[0, 1].set_title('2nd')
axs[1, 0].set_title('3rd')
axs[1, 1].set_title('4th')

fig.suptitle('I am Four Line Plots')
plt.xlabel('I am X')
plt.ylabel('I am Y')
plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

--> Bar plots are useful for visualizing categorical data or discrete values, such as counts or percentages. They can be used to compare different categories or groups of data, and can also be used to show changes over time or across different groups.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.xlabel('Compnies')
plt.ylabel('Profit in Cr')
plt.show()

plt.barh(company,profit)
plt.xlabel('Compnies')
plt.ylabel('Profit in Cr')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

--> Box plots are useful for identifying the central tendency, variability, and distribution of a dataset. They can be used to compare multiple datasets or to identify any unusual observations or outliers in a dataset.



```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

a=[box1,box2]

plt.boxplot(a)
plt.title('BoxPlot')
plt.xlabel('x')
plt.ylabel('Y')
plt.xticks([1,2],['Box1','Box2'])
plt.show()
```
